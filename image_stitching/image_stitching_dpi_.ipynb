### Implemente ```image stitching``` on images ['ic1.png', 'ic2.png', 'ic3.png', 'ic4.png'] to generate one panorama image 'ic.png'.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
read_images = ['./ic1.jpg', './ic2.jpg', './ic3.jpg', './ic4.jpg']
imgs = [cv2.imread(img) for img in read_images]


# Definir o objeto stitcher
stitcher = cv2.Stitcher_create()

# Criar a imagem panor√¢mica
status, pano = stitcher.stitch(imgs)
````

````python
saving = cv2.imwrite('ic.png', cv2.cvtColor(pano, cv2.COLOR_BGR2RGB))
result = cv2.imread('./ic.png')
````

````python
plt.imshow(result)
````
