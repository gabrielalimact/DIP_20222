```python
img = cv.imread('Albert_Einstein_Head.jpg')
```

```python
img2 = cv.imread('cuba.jpeg')
```

```python
# (1) Crop and flip an image using Numpy array indexing.
crop_img = img[1000:2500, 2000:3500]
flip_img = np.fliplr(img)

fig=plt.figure(figsize=(15,10))
plt.subplot(1, 3, 1) # row 1, col 2 index 1
plt.imshow(img)
plt.title("Original")
plt.subplot(1, 3, 2) # index 2
plt.imshow(crop_img)
plt.title("cropped image")
plt.subplot(1, 3, 3) # index 3
plt.imshow(flip_img)
plt.title("flipped image")
```

```python
# (2) Implement image translation using Numpy and OpenCV.
x = 200 # number of pixels to shift in x 
y = 500 # number of pixels to shift in y
T = np.float32([[1, 0, x], [0, 1, y]]) # translation matrix

translated_img = cv.warpAffine(img, T, (img.shape[1], img.shape[0]))

fig=plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.imshow(img)
plt.title("Original")
plt.subplot(1, 2, 2) # index 2
plt.imshow(translated_img)
plt.title("translated image with {} pixels in x and {} in y".format(x, y))
```

```python
# (3) Implement image rotation using Numpy and OpenCV.
angle = 45
center = (img.shape[1]//2, img.shape[0]//2)
R = cv.getRotationMatrix2D(center, angle, 1.0) # rotetion matrix

rotated_img = cv.warpAffine(img, R, (img.shape[1], img.shape[0]))

fig=plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.imshow(img)
plt.title("Original")
plt.subplot(1, 2, 2) # index 2
plt.imshow(rotated_img)
plt.title("rotated image with {} degree".format(angle))
```

```python
# (4) Implement image resizing using OpenCV.
w, h = 500, 500
dim = (w, h)

resized_img = cv.resize(img, dim, interpolation=cv.INTER_AREA)


fig=plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.imshow(img)
plt.title("Original")
plt.subplot(1, 2, 2) # index 2
plt.imshow(resized_img)
plt.title("resized image to {}x{}".format(w, h))
```

```python
# (5) Implement bitwise operations: AND, OR, XOR.
resized_img1 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
resized_img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

# resize images to match 
w, h = 500, 500
dim = (w, h)
resized_img1 = cv.resize(img, dim, interpolation=cv.INTER_AREA)
resized_img2 = cv.resize(img2, dim, interpolation=cv.INTER_AREA)


# and
bit_and = cv.bitwise_and(resized_img1, resized_img2)
# or
bit_or = cv.bitwise_or(resized_img1, resized_img2)
# xor
bit_xor = cv.bitwise_xor(resized_img1, resized_img2)

fig=plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.imshow(resized_img1)
plt.title("Original 1")
plt.subplot(1, 2, 2) # index 2
plt.imshow(resized_img2)
plt.title("Original 2")
plt.show()
print()
fig=plt.figure(figsize=(15,10))
plt.subplot(1, 3, 1) # row 1, col 2 index 1
plt.imshow(bit_and)
plt.title("AND")
plt.subplot(1, 3, 2) # index 2
plt.imshow(bit_or)
plt.title("OR")
plt.subplot(1, 3, 3) # index 3
plt.imshow(bit_xor)
plt.title("XOR")
plt.show()
```
